## Problem 1: Mapping Atlantis' Hidden Chambers


```python
def map_chambers(sections):
    if type(sections) == str: return sections
    if len(sections) == 1:
        return sections[0]
    return {sections[0]: map_chambers(sections[1])}
```

```python
sections = ["Atlantis", ["Coral Cave", ["Pearl Chamber"]]]
print(map_chambers(sections))
```

## Problem 2: Finding the Longest Sequence of Trident Gems


```python
def longest_trident_sequence(gems):
    # def helper(count, next):
    #     if next == len(gems): return 0
    #     # print(gems[next])
    #     if gems[next] - gems[next - 1] == 1:
    #         return 1 + helper(count, next + 1)
    #     else:
    #         return helper(count, next + 1)
    # return helper(0, 1)

    def helper(index, curlen, maxlen):
        if index == len(gems):
            return curlen
        
        if gems[index] - gems[index - 1] == 1:
            curlen += 1
        else:
            curlen = 1

        maxlen = max(maxlen, curlen)
        return helper(index + 1, curlen, maxlen)
    return helper(0, 1, 1)
```

```python
print(longest_trident_sequence([1, 2, 3, 2, 3, 4, 5, 6]))
print(longest_trident_sequence([5, 10, 7, 8, 1, 2]))
```

## Problem 3: Last Building Standing

```python
# iterative
from collections import deque


def find_last_building(n, k):
    q = deque([i for i in range(1, n + 1)])
    while len(q) > 1:
        for _ in range(k - 1):
            q.append(q.popleft())
        q.popleft()
    return q[0]
```

## Problem 5: Merging Missions II


```python
class Node:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

# For testing
def print_linked_list(head):
    current = head
    while current:
        print(current.value, end=" -> " if current.next else "\n")
        current = current.next

def merge_missions_iterative(mission1, mission2):
    temp = Node()  # Temporary node to simplify the merging process
    tail = temp

    while mission1 and mission2:
        if mission1.value < mission2.value:
            tail.next = mission1
            mission1 = mission1.next
        else:
            tail.next = mission2
            mission2 = mission2.next
        tail = tail.next

    # Attach the remaining nodes, if any
    if mission1:
        tail.next = mission1
    elif mission2:
        tail.next = mission2

    return temp.next  # Return the head of the merged linked list
```

```python
mission1 = Node(1, Node(2, Node(4)))
mission2 = Node(1, Node(3, Node(4)))

print_linked_list(merge_missions_iterative(mission1, mission2))
```

## Problem 6: Decoding Ancient Atlantean Scrolls


```python
def decode_scroll(scroll):
    stack = []
    current_string = ""
    current_num = 0
    
    for char in scroll:
        if char.isdigit():
            # Build the number (could be more than one digit)
            current_num = current_num * 10 + int(char)
        elif char == '[':
            # Push the current number and current string to the stack
            stack.append((current_string, current_num))
            # Reset the current string and number
            current_string = ""
            current_num = 0
        elif char == ']':
            # Pop the last string and number from the stack
            prev_string, num = stack.pop()
            # Repeat the current string num times and add it to the previous string
            current_string = prev_string + current_string * num
        else:
            # Regular character, just add it to the current string
            current_string += char
    
    return current_string

def decode_scroll_recursive(scroll):
    def decode(pos):
        curstr = ''
        curnum = 0
        while pos < len(scroll):
            char = scroll[pos]
            if char.isdigit():
                curnum = curnum*10 + int(char)
            elif char == ']':
                return curstr, pos
            elif char == '[':
                decode_str, pos = decode(pos + 1)
                curstr += decode_str * curnum
                curnum = 0
            else:
                curstr += char
            pos += 1

        return curstr, pos
    res, _ = decode(0)
    return res
```

```python
scroll = "3[Coral2[Shell]]"
print(decode_scroll_recursive(scroll))

scroll = "2[Poseidon3[Sea]]"
print(decode_scroll_recursive(scroll))
```

## Problem 1: Extra Treats


```python
from collections import deque


def predictAdoption_victory(votes):
    cats = deque()
    dogs = deque()
    n = len(votes)

    for i, v in enumerate(votes):
        if v == 'C':
            cats.append(i)
        else:
            dogs.append(i)
    

    while cats and dogs:
        c = cats.popleft()
        d = dogs.popleft()

        if c < d:
            cats.append(c + n)
        else:
            dogs.append(d + n)

    return "Cat Lovers" if cats else "Dog Lovers"
```

```python
print(predictAdoption_victory("CD")) 
print(predictAdoption_victory("CDD")) 
print(predictAdoption_victory("DDDDDDDCCCDCDCCCCC")) 
```

## Problem 2: Pair Up Animals for Shelter


```python
def pair_up_animals(discomfort_levels):
    discomfort_levels.sort()
    left = 0
    right = len(discomfort_levels) - 1
    maxdiscomfort = 0

    while left < right:
        maxdiscomfort = max(maxdiscomfort, discomfort_levels[left] + discomfort_levels[right])
        left += 1
        right -= 1
    return maxdiscomfort
```

```python
print(pair_up_animals([3,5,2,3]))  
print(pair_up_animals([3,5,4,2,4,6])) 
```

## Problem 3: Rearrange Animals and Slogans


```python
def rearrange_animal_names(s):
    stack = []

    for i in s:
        if i != ')':
            stack.append(i)
        else:
            tempstack = []
            while True:
                x = stack.pop()
                if x == '(':
                    break
                tempstack.append(x)
            stack.extend(tempstack)
        # else:
        #     temp = []
        #     while stack and stack[-1] != '(':
        #         temp.append(stack.pop())
        #     stack.pop()
        #     stack.extend(temp)
    print(''.join(stack))
```

## Problem 4: Append Animals to Include Preference


```python
def append_animals(available, preferred):
    left = 0
    right = 0

    while left < len(available) and right < len(preferred):
        if available[left] == preferred[right]:
            right += 1
        # if right == len(preferred) - 1:
        #     return 0
        # else:
        left += 1

    return len(preferred) - right
```

```python
print(append_animals("catsdogs", "cows")) 
print(append_animals("rabbit", "r")) 
print(append_animals("fish", "bird"))
```

## Problem 5: Group Animals by Habitat


```python
def group_animals_by_habitat(habitats):
    last = {ch: i for i, ch in enumerate(habitats)}
    start = end = 0
    result = []
    maxidx = 0

    for idx, char in enumerate(habitats):
        end = max(end, last[char])
        if idx == end:
            result.append(end - start + 1)
            start = idx + 1
    return result

```

## Problem 6: Validate Animal Sheltering Sequence


```python
def validate_shelter_sequence(admitted, adopted):
    stack = []
    j = 0

    for animal in admitted:
        stack.append(animal)
        while stack and j < len(adopted) and stack[-1] == adopted[j]:
            stack.pop()
            j += 1
    return j == len(adopted)
```

```python
print(validate_shelter_sequence([1,2,3,4,5], [4,5,3,2,1]))
print(validate_shelter_sequence([1,2,3,4,5], [4,3,5,1,2])) 
```
